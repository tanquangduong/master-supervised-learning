# Required package

```python
import pandas as pd
import sklearn
```

# Importing data

```python
dataset = pd.read_csv('./datasets/Social_Network_Ads.csv')
dataset.head()
```

# Feature selection

```python
dataset = dataset.drop(columns=['User ID'])
dataset.head()
```

# One-hot encoding

```python
from sklearn import preprocessing
# training OneHotEncoder
enc = preprocessing.OneHotEncoder()
enc.fit(dataset.iloc[:,[0]])
onehotlabels = enc.transform(dataset.iloc[:,[0]]).toarray()
print(onehotlabels)
```

```python
# create dataframe for gender from onehotlabels
genders = pd.DataFrame({'Female': onehotlabels[:, 0], 'Male': onehotlabels[:, 1]})
genders.head()
```

```python
# concatenate the final dataframe
result = pd.concat([genders,dataset.iloc[:,1:]], axis=1, sort=False)
result.head(5)
```

# Specifying the features and label

```python
y=result['Purchased']
X=result.drop(columns=['Purchased'])
```

# Dividing the dataset into testing and training portions

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

# Scaling the features

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Call Logistic Regression algorithm

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

# Evaluation

```python
from sklearn import metrics
# Predicting the Test set results
y_pred = classifier.predict(X_test)
cm = metrics.confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy= metrics.accuracy_score(y_test,y_pred)
recall = metrics.recall_score(y_test,y_pred)
precision = metrics.precision_score(y_test,y_pred)
print(accuracy,recall,precision)
```
